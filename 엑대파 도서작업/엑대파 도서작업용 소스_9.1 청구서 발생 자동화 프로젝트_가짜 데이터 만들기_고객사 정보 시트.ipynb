```python
import random
import string
import pandas as pd
from korean_name_generator import namer

def make_random_string():
    str_length = random.randrange(4,9)
    string_pool = string.ascii_lowercase
    result = ""
    for i in range(str_length):
        result += random.choice(string_pool)
    return result

for i in range(len(df)):
    this_name = namer.generate(True).replace("혀", "현")
    df._set_value(i, "담당자", this_name)
    df._set_value(i, "담당자 이메일","{}@{}.kr".format(make_random_string(), make_random_string()))
df.to_excel("sample_data/고객사 명단.xlsx", index=False)
df
```
