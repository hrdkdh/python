```python
import pandas as pd

df_ph = pd.read_excel("c:/Users/POSCOUSER/Desktop/지역아동센터_포항(크롤링).xlsx", index_col=0)
df_kw = pd.read_excel("c:/Users/POSCOUSER/Desktop/지역아동센터_광양(크롤링).xlsx", index_col=0)
df_old = pd.read_excel("c:/Users/POSCOUSER/Desktop/210316_지역아동센터 현황(최종)_포항.xlsx")
```

```python
df_old
df_old_replace_arr = {
    "해오름공부방": "해오름",
    "한울공부방": "한울",
    "엘림": "포항엘림",
    "연일참사랑": "참사랑",
    "양포행복한홈스쿨": "양포행복한"
}
for rp_org, rp_new in df_old_replace_arr.items():
    df_old["시설명"] = df_old["시설명"].replace(rp_org, rp_new)
df_old
```

```python
# df_ph.replace({"센터명" : {"(.*)지역아동센터(.*)", r"\1\2"}}, regex=True, inplace=True)
df_ph["센터명2"] = df_ph["센터명"].replace("(.*)지역아동센터(.*)", r"\1\2", regex=True)
df_ph = df_ph[["센터명", "센터명2", "주소", "홈페이지", "전화", "팩스", "이메일"]]
add_col = ["그룹사 연계", "정원", "현원", "종사자수", "신고일", "면적(㎡)", "전화2"]
for col in add_col:
    df_ph[col] = ""

df_ph
```

```python
checked_cnt = 0
unchecked_cnt = 0
unchecked_list = []
for i_new in range(len(df_ph)):
    checked = False
    for i_old in range(len(df_old)):
        if df_ph.iloc[i_new, 1] == df_old.iloc[i_old, 0]:
            checked_cnt += 1
            checked = True
            df_ph.iloc[i_new, 7] = df_old.iloc[i_old, 12]
            df_ph.iloc[i_new, 8] = df_old.iloc[i_old, 2]
            df_ph.iloc[i_new, 9] = df_old.iloc[i_old, 3]
            df_ph.iloc[i_new, 10] = df_old.iloc[i_old, 4]
            df_ph.iloc[i_new, 11] = df_old.iloc[i_old, 5]
            df_ph.iloc[i_new, 12] = df_old.iloc[i_old, 6]
            df_ph.iloc[i_new, 13] = df_old.iloc[i_old, 14]
            break
    if checked == False:
        unchecked_cnt += 1
        unchecked_list.append(df_ph.iloc[i_new, 1])

print("반영 {}건, 미반영 {}건".format(checked_cnt, unchecked_cnt))
print(unchecked_list)
df_ph.to_excel("c:/Users/POSCOUSER/Desktop/지역아동센터_포항(크롤링)_후처리 완료.xlsx")
df_ph
```
